```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from random import randint
import scipy.optimize as optimization
```

```python
discriminator = [-5, -10, -15, -20, -30, -40, -50, -60, -80, -100, -120] #mv
counts = [11026, 13379, 12323, 11123, 8607, 5382, 2707, 1011, 78, 14, 9]

counts_bkgrn = [1779, 2350, 2143, 1923, 1444, 1051, 556, 298, 25, 6, 3]
counts_bkgrn_10 = []

for i in counts_bkgrn:
    counts_bkgrn_10.append(i/10.0)
 
S_N = []
for i in range(11):
    S_N.append(counts[i]/counts_bkgrn_10[i])
```

```python
plt.plot(discriminator, S_N, '^')
e, d, c, b, a = pylab.polyfit(discriminator, S_N, 4)
m, g = pylab.polyfit(discriminator, S_N, 1)

x = np.arange(-120, 0, 1)
y = []
z = []
for i in x:
    y.append(a + b*i + c*(i**2) + d*(i**3) + e*(i**4))
    z.append(g + m*i)
plt.plot(x, y, 'g')
plt.plot(x, z, 'r')
plt.show()
print "peak is at approximitely -30mv"
```
