```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#lead
plates = [0, 5.4, 10.6, 15.99, 21.27, 26.48, 31.71, 42.07]
N_gross = [9932, 7277, 5359, 3771, 2796, 2031, 1430, 805]
N_net = [7664, 5849, 4435, 2847, 2334, 1737, 1136, 721]
percent_uncer = [1.44, 1.60, 1.79, 2.41, 2.45, 2.78, 3.66, 4.14]

N_back = []
uncer = []
for i in range(8):
    N_back.append(N_gross[i] - N_net[i])
    uncer.append(N_net[i]*percent_uncer[i])

logN_net = []
del_logN_net = []
for i in range(8):
    logN_net.append(np.log(N_net[i]))
    del_logN_net.append(uncer[i]/N_net[i])
    
plt.errorbar(plates, logN_net, del_logN_net, 0, '.')
plt.xlabel('Total Height of Lead Plates (mm)')
plt.ylabel('Log(Net Counts) (Counts per 200 seconds)')
plt.show()

x = plates
y = logN_net
w = 0
wx = 0
wy = 0
wxy = 0
wx2 = 0
for i in range(8):
    w += 1.0/(del_logN_net[i]**2)
    wx += x[i]/(del_logN_net[i]**2)
    wy += y[i]/(del_logN_net[i]**2)
    wxy += (x[i]*y[i])/(del_logN_net[i]**2)
    wx2 += (x[i]**2)/(del_logN_net[i]**2)

wwx2 = 0
for i in range(8):
    wwx2 += wx2/(del_logN_net[i]**2)
    
delta += (wwx2) - (wx**2)
    

A = ((wx2 * wy) - (wx * wxy))/delta
B = ((w * wxy) - (wx * wy))/delta

Fit_logI = []
for i in x:
    Fit_logI.append(A + B*i)
    
plt.plot(x, Fit_logI, 'g')
plt.show()

sig_A = np.sqrt((wx2)/delta)
sig_B = np.sqrt((w/delta))

print 'Intercept:', A, '+/-', sig_A
print 'Slope:', B, '+/-', sig_B

plt.errorbar(plates, logN_net, del_logN_net, 0, '.')
plt.plot(x, Fit_logI, 'g')
plt.xlabel('Total Height of Lead Plates (mm)')
plt.ylabel('Log(Net Counts) (Counts per 200 seconds)')
plt.show()
```
